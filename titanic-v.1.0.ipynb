# **Explore A Pattern**

```python
women = dataTrain.loc[dataTrain.Sex == 'female']['Survived']
rate_women = sum(women)/len(women)

print(f'women who survived : {rate_women}')
```

```python
man = dataTrain.loc[dataTrain.Sex == 'male']['Survived']
rate_man = sum(man)/len(man)

print(f'man who survived : {rate_man}')
```

# **Looking and Checking the Data Again**

# **Modelling**

```python
from sklearn.ensemble import RandomForestClassifier


# labels
y = dataTrain['Survived']
# feature
features = ['Pclass', 'Sex', 'SibSp', 'Parch']

# One Hot Encod
X = pd.get_dummies(dataTrain[features])
testX = pd.get_dummies(dataTest[features])

# Model
model = RandomForestClassifier(
    n_estimators=100,
    random_state=1,
    max_depth=5,
    min_samples_split=4,
    min_samples_leaf=1,
)

# Fit 
model.fit(X, y)
```

# **Predictions**

```python
prediction = model.predict(testX)
```

```python
print(prediction)
```

```python
output = pd.DataFrame({
    'PassengerId': dataTest.PassengerId,
    'Survived': prediction})

output.to_csv('MySubmissions', index=False)
print('Your mission was succesfully saved!')
```
